## Steps:
    1. Load Boston Dataset
    2. Introduce new features as x1_x1, x1_x2 and so on
    3. Compare the Scores

```python
import numpy as np
import pandas as pd
from sklearn import datasets
boston_dataset = datasets.load_boston()
data = boston_dataset.data
target = boston_dataset.target
df = pd.DataFrame(data)
df.columns = boston_dataset.feature_names
num_features = len(df.columns)
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
def train_test_score (X_train, X_test, Y_train, Y_test) :
    lin_regr = LinearRegression()
    lin_regr.fit(X_train, Y_train)
    train_score = lin_regr.score(X_train, Y_train)
    test_score = lin_regr.score(X_test, Y_test)
    return train_score, test_score
```

```python
from sklearn import model_selection
train_scores = []
test_scores = []
X = data
Y = target

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y)
new_num_features = df.shape[1]
train_score, test_score = train_test_score(X_train, X_test, Y_train, Y_test)
train_scores.append(train_score)
test_scores.append(test_score)


prev_df = df.copy()
col_names = boston_dataset.feature_names
for i in range(num_features) :
    for j in range(i, num_features) :
        new_df = prev_df.copy()
        col_i_name = col_names[i]
        col_j_name = col_names[j]
        new_df[col_i_name + "_" + col_j_name] = prev_df[col_i_name] * prev_df[col_j_name]
        X = new_df.values
        X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, random_state = None)
        #print(new_df.shape)
        new_num_features = df.shape[1]
        train_score, test_score = train_test_score(X_train, X_test, Y_train, Y_test)
        train_scores.append(train_score)
        test_scores.append(test_score)
        prev_df = new_df
        
```

```python
# with random state = 1 in train_test_split
#print(train_scores, test_scores)
nf = np.arange(13,105)
import matplotlib.pyplot as plt
plt.figure(figsize = (16, 9))
plt.xlabel("Number of Features")
plt.ylabel("Score")
plt.plot(nf, train_scores, color ="b", label = "Train Score", marker = ".")
plt.plot(nf, test_scores, color ="r", label = "Test Score", marker = ".")
plt.legend()
plt.grid()
plt.show()
```
