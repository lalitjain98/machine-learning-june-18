```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
oli = datasets.fetch_olivetti_faces()
```

```python
olive = oli
```

```python
olive.keys()
```

```python
from pprint import pprint
print(olive.DESCR)
```

```python
olive['images'].shape
```

```python
olive['data'].shape
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(olive['images'][i], cmap = plt.cm.bone)
plt.show()
```

```python
X = olive.data
Y = olive.target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0)
```

```python
pca = PCA()
pca.fit(X_train)
```

```python
pca.components_.shape
```

```python
total = sum(pca.explained_variance_)
k = 0
c_sum = 0
while c_sum/total < 0.95 :
    c_sum += pca.explained_variance_[k]
    k += 1
k
```

```python
pca = PCA(n_components = k, whiten = True)
transformed_data = pca.fit_transform(X_train)
transformed_data.shape
```

```python
x_approx = pca.inverse_transform(transformed_data)
x_approx.shape
```

```python
x_approx_images = x_approx.reshape((300, 64, 64))
```

```python

fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(x_approx_images[i], cmap = plt.cm.bone)
plt.show()
```

```python
eigen_vecs = pca.components_
```

```python
eigen_faces = eigen_vecs.reshape((107, 64, 64))
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1)
    ax.imshow(eigen_faces[i], cmap = plt.cm.bone)
plt.show()
```

```python
X_train_transformed = transformed_data
model = LogisticRegression()
model.fit(X_train_transformed, Y_train)
model.score(X_train_transformed, Y_train)
```

```python
X_test_transformed = pca.transform(X_test)
```

```python
model.score(X_test_transformed, Y_test)
```

```python
Y_train_pred = model.predict(X_train_transformed)
Y_test_pred = model.predict(X_test_transformed)
```

```python
print(confusion_matrix(Y_train_pred, Y_train))
print(confusion_matrix(Y_test_pred, Y_test))
```

```python
print(classification_report(Y_train_pred, Y_train))
print(classification_report(Y_test_pred, Y_test))
```
