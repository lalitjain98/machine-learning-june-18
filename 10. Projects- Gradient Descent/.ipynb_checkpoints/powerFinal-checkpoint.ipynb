```python
import numpy as np
```

```python
def cost(result,m,train):
    c=0
    for i in range(len(train)):
        c+= (result[i][0]-(m*train[i]).sum())**2
    return c/len(train)
```

```python
def fit(train_x,train_y,alpha,itr):
    slope=np.array([0 for i in range(len(train_x[0]))])
    m=np.array([0 for i in range(len(train_x[0]))])
    for i in range(itr):
        slp=0
        for k in range(len(train_x[0])):
            for j in range(len(train_x)):
                slp+=(-2*train_x[j][k]/(len(train_x)))*(train_y[j][0]-(m*train_x[j]).sum())
            slope[k]=slp
        m=m-alpha*slope
        print("cost",i+1,":",cost(train_y,m,train_x),"    ",m)
    return m          
    
```

```python
def predict(inp,m):
    result=np.array([0 for i in range(len(inp))])
    result=result.reshape(len(result),1)
    m=m.reshape(len(m),1)
    for i in range(len(inp)):
        result[i]=(inp[i]*m).sum()
    return result
```

```python
def loadAndShape():
    train_data=np.loadtxt('training_ccpp_x_y_train.csv.csv',delimiter=',')
    train_x=train_data[:,0:len(train_data[0])-1]
    train_y=train_data[:,len(train_data[0])-1:len(train_data[0])]
    test_data=np.loadtxt('test_ccpp_x_test.csv.csv',delimiter=',')
    test_x=test_data
    ap=np.array([1 for i in range(len(test_x))])
    ap=ap.reshape(len(test_x),1)
    test_x= np.append(test_x,ap,axis=1)
    
    ap=np.array([1 for i in range(len(train_x))])
    ap=ap.reshape(len(train_x),1)
    train_x= np.append(train_x,ap,axis=1)
    return train_x,train_y,test_x        
```

```python
def manager():
    train_x,train_y,test_x=loadAndShape()
    alpha=0.0000001
    itr=1000
    m=fit(train_x,train_y,alpha,itr)
    finalResult=predict(test_x,m)
   # print(finalResult.shape)
    np.savetxt("prediction_power.csv", finalResult, delimiter=",")   
    return 
```

```python
manager()
```
