```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
```

```python
plt.scatter(x, y) 
```

```python
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
# plot two different lines
import numpy as np
x2 = np.arange(1, 5)
y2 = x2 * 3
plt.plot(x,y)
plt.plot(x2, y2)
plt.show()
```

```python
plt.scatter(x,y)
plt.scatter(x2, y2)
plt.show()
```

```python
# a third arg may also be passed
# consists of two characters
# first character is color : 
# r for red, b for blue, g for green etc
# second character is shape
# - for line plot, -- for dashed line plot
# ^ for triangles, o for circles etc
plt.plot(x,y, "r-")
plt.plot(x2, y2, "g^")
plt.plot(x2, y2, "g--")
plt.show()
```

```python
# if only one list is passes as arg to plot or scatter function
# plt automatically assumes x = range(n)

l = (np.arange(20)) ** 3
plt.plot(l,"r--")
plt.show()
```

```python
# customizing graphs

l = (np.arange(20)) ** 3
plt.plot(l, color = "black", marker = "p")
plt.show()
```

```python

l = (np.arange(20)) ** 3
plt.plot(l, color = "black", marker = "o")
plt.show()
```

```python

l = (np.arange(20)) ** 3
plt.plot(l, color = "black", linewidth = 5)
plt.show()
```

```python
# Add title, legends, axis labels
y1 = (np.arange(20)) ** 3
y2 = (np.arange(20)) ** 2
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plot for Different Functions")
plt.plot(y1, color = "blue", linewidth = 2, label = "X ^ 3")
plt.plot(y2, color = "green", linewidth = 2, label = "X ^ 2")
plt.show()
```

```python
y1 = (np.arange(0, 10, 0.2)) ** 3
y2 = (np.arange(0, 10, 0.2)) ** 2
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plot for Different Functions")
plt.plot(y1, color = "blue", linewidth = 2, label = "X ^ 3")
plt.plot(y2, color = "green", linewidth = 2, label = "X ^ 2")
# actually show the legends passed in plt functions
plt.legend() 
plt.show()
```

```python
# alter the axes
# show the grid
y1 = (np.arange(0, 10, 0.2)) ** 3
y2 = (np.arange(0, 10, 0.2)) ** 2

plt.axis([0, 50, 0 ,100])
plt.grid()

plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plot for Different Functions")
plt.plot(y1, color = "blue", linewidth = 2, label = "X ^ 3")
plt.plot(y2, color = "green", linewidth = 2, label = "X ^ 2")
# actually show the legends passed in plt functions
plt.legend() 
plt.show()
```

```python
# add text at some position (x, y) on graph

# alter the axes
# show the grid
y1 = (np.arange(0, 10, 0.2)) ** 3
y2 = (np.arange(0, 10, 0.2)) ** 2

plt.text(15, 70, "X^3", fontsize = "12")
plt.text(30, 30, "X^2", fontsize = "12")

plt.axis([0, 50, 0 ,100])
plt.grid()

plt.ylabel("Y")
plt.xlabel("X")
plt.title("Plot for Different Functions")
plt.plot(y1, color = "blue", linewidth = 2, label = "X ^ 3")
plt.plot(y2, color = "green", linewidth = 2, label = "X ^ 2")
# actually show the legends passed in plt functions
plt.legend() 
plt.show()
```

```python
# Creating Pie Graphs
sizes = [3, 4, 5, 6]
plt.pie(sizes)
plt.show()
```

```python
# plt.axis("equal") for perfect circle

sizes = [3, 4, 5, 6]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
sizes = [3, 4, 5, 6]
colors = ["yellow", "orange", "red", "lime"]
plt.pie(sizes, colors = colors)
plt.axis("equal")
plt.show()
```

```python
# Add Labels

sizes = [3, 4, 5, 6]
labels = ["A", "B", "C", "D"]
colors = ["yellow", "orange", "red", "lime"]
plt.pie(sizes, colors = colors, labels = labels)
plt.axis("equal")
plt.show()
```

```python
# Add title
# Add Percentages of area
# parameter is autopct
# either pass a function or format
# %% is kind of like an escape seq for % symbol
sizes = [3, 4, 5, 6]
labels = ["A", "B", "C", "D"]
colors = ["yellow", "orange", "red", "lime"]

plt.title("A Pie Graph")

plt.pie(sizes, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()


plt.pie(sizes, colors = colors, labels = labels, autopct = "%.2f%%")
plt.axis("equal")
plt.show()
```

```python
sizes = [3, 4, 5, 6]
labels = ["A", "B", "C", "D"]
colors = ["yellow", "orange", "red", "lime"]

plt.title("A Pie Graph")
explode = [0.1, 0, 0, 0]
plt.pie(sizes, explode = explode, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()

explode = [0.5, 0, 0, 0]
plt.pie(sizes, explode = explode, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()

explode = [1, 0, 0, 0]
plt.pie(sizes, explode = explode, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()

explode = [0.2, 0.1, 0.1, 0.1]
plt.pie(sizes, explode = explode, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()
```

```python
# startangle is the angle from where 
# to start drawing the pie graph

# pass counterclock = True 
# for elements to appear in 
# counterclockwise manner

sizes = [3, 4, 5, 6]
labels = ["A", "B", "C", "D"]
colors = ["yellow", "orange", "red", "lime"]

plt.title("A Pie Graph")
explode = [0.1, 0, 0, 0]
plt.pie(sizes, startangle = 100, counterclock = True, explode = explode, colors = colors, labels = labels, autopct = "%.2f")
plt.axis("equal")
plt.show()
```
