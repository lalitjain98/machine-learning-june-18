```python
# manipulating dataframes

#drop removes the row whose index is passed
#and returns the altered dataframe

# to do it inplace pass inplace = True

"""
first argument passed is the 
actual index of the row to delete
every row has an index
so you cannot delete the 0th row more than once
"""

df = iris.copy()
print(df.head())
df2 = df.drop(2)
print(df2.head())
print(df.head())

#change back to initial df
df2 = df
df2.drop(3, inplace = True)
print(df2.head())
```

```python
# print row numbers/indices
print(df.index)


# df.index is list of indices of rows
# (that is list of row names)
# so to remove ith row
# pass df.index[i] as index for drop function


# eg to remove 0th row
df.drop(df.index[0])
```

```python
df[0] > 5
```

```python
df.columns = ['sl', 'sw', 'pl', 'pw', "flower_type"]
df[df.flower_type == 'Iris-virginica'].describe()
```

```python
# iloc works by position
# loc works by row index name

print(df.head())
print(df.iloc[0])
print(df.loc[5])

# add a row to dataframe

df.loc[df.shape[0] + 2] = [1.5, 2.7, 3.2, 4.8, "Iris-setosa"]
print(df.tail())
```

```python

df.drop(df.index[1:10])

# now we would want to reset indices
print(df.reset_index())
print(df.head())
# pass drop = True to remove previous index column
print(df.reset_index(drop = True))
print(df.head())
# to make changes to the df itself pass inplace = True
print(df.reset_index(drop = True, inplace = True))
print(df.head())
# to remove column pass axis = 1
# inplace works here too
```

```python
df.drop('sl', axis = 1)
```
