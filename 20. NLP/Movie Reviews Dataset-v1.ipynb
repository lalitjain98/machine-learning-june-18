```python
documents = []
for cat in movie_reviews.categories() :
    for id in movie_reviews.fileids(cat):
        documents.append((movie_reviews.words(id), cat))
print(documents[:5])
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag) :
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
import random
```

```python
random.shuffle(documents)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import time
def clean_review(review, stops) :
    words = []
    for word in review:
        if word.lower() not in stops :
            pos = pos_tag(word)
            clean_word = lemmatizer.lemmatize(word, pos = get_simple_pos(pos[0][1]))
            words.append(clean_word.lower())
    return words
```

```python
import string
from nltk.corpus import stopwords
stops = list(stopwords.words('english'))
stops += (list(string.punctuation))
stops
```

```python
import time

clean_reviews = []
i = 0
st = time.time()
for review, cat in documents :
    i += 1
    clean_reviews.append((clean_review(review, stops), cat))
    if (i+1) % 100 == 0 :
        et = time.time()
        print(i, "Reviews Cleaned in ", round(et-st), "sec")
        st = time.time()
print("Time", round((et-st)), "sec")
```

```python
only_reviews = [' '.join(review[0]) for review in clean_reviews]
only_classes = [review[1] for review in clean_reviews]
```

```python
only_classes[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df["review"] = only_reviews
df["class"] = only_classes
df.head()
```

```python
df.to_csv("clean_reviews.csv", index = False)
```

```python
train_data = df.values[:1500]
test_data = df.values[1500:]
```

```python

all_words = []
for review in df['review'].values :
    all_words += review.split()
```

```python
import nltk
```

```python
freq_dist = nltk.FreqDist(all_words)
```

```python
most_common = freq_dist.most_common(3000)
```

```python
features = [ i[0] for i in most_common]
features[:10]
```

```python
def get_feature_dictionary(words) :
    current_features = {}
    words_set = set(words)
    for word in features :
        current_features[word] = word in words_set
    return current_features
```

```python
training_data = [(get_feature_dictionary(review.split()), cat) for review, cat in train_data]
```

```python
testing_data = [(get_feature_dictionary(review.split()), cat) for review, cat in test_data]
```
