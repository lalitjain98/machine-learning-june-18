```python
import numpy as np
import pandas as pd
```

```python
dataset_df = pd.read_csv("20_newsgroups_dataset.csv")
dataset_df.describe()
```

```python
X_train_df = dataset_df.drop('target', axis = 1) 
Y_train_df = dataset_df['target']
```

```python
X_train_df.describe()
```

```python
Y_train_df.describe()
```

```python
words = X_train_df.columns
print(words)
possible_classes = list(set(Y_train_df.values))
print(possible_classes)
```

```python

def predict_single(X, model) :
    possible_classes = model['possible_classes']
    num_words = model["vocabulary_size"]
    max_class = None
    max_class_prob = -np.inf
    
    for y in possible_classes :
        prob_Y_equals_y = np.log(model[y]["class_count"]/model["total_data"])
        prob_X_equals_x_given_Y_equals_y = 0
        for f in range(num_words) :
            prob_X_equals_x_given_Y_equals_y += X[f] * np.log(((model[y][f]/model[y]["total_words"]) + 1)/num_words)
        if prob_X_equals_x_given_Y_equals_y > max_class_prob :
            max_class_prob = prob_X_equals_x_given_Y_equals_y
            max_class = y
    return max_class
            
def predict(X, model) :
    #if not isinstance(X[0], list) :
    #    return None #predict_single(X, model)
    Y_pred = []
    for x in X :
        Y_pred.append(predict_single(x, model))
    return Y_pred

def fit(X, Y) :
    num_words = len(X[0])
    possible_classes = list(set(Y))
    model = {}
    model["total_data"] = len(Y)
    model["vocabulary_size"] = num_words
    model["possible_classes"] = possible_classes
    for y in possible_classes :
        y_dict = {i : X[Y == y, i].sum() for i in range(num_words)}
        y_dict['total_words'] = sum(y_dict.values())
        y_dict['class_count'] = sum(Y == y)
        #print(y_dict['total_count'])
        
        model[y] = y_dict
    return model
```

```python
print(model)
```

```python
model = fit(X_train_df.values, Y_train_df.values)
```

```python
import time
st = time.time()
Y_pred = predict(X_train_df.values, model)
et = time.time()

print("Runtime:", (int((et-st)*1000))/1000)
print(Y_pred)
```

```python
training_time = (et-st)/60
print(training_time)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from pprint import pprint
pprint(confusion_matrix(Y_pred, Y_train_df.values))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_pred, Y_train_df.values))
```
